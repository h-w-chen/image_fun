```python
# based on https://www.kaggle.com/datasets/adityajn105/flickr8k
from keras.applications import ResNet50
from keras.applications.regnet import preprocess_input
from keras.models import Model, load_model
```

```python
model = ResNet50(weights='imagenet', input_shape=(224,224,3))
model.summary()
```

```python
model_n = Model(model.input, model.layers[-2].output)
```

```python
model_n.summary()
```

```python
from keras_preprocessing import image
import numpy as np
```

```python
def preprocess_img(img):
    img = image.load_img(img, target_size=(224,224,))
    img = image.img_to_array(img)
    img = np.expand_dims(img, axis=0)
    img = preprocess_input(img)
    return img

def decode_img(img):
    img = preprocess_img(img)
    feature_v = model_n.predict(img)
    feature_v = feature_v.reshape((-1,))
    return feature_v
```

```python
f_v = decode_img('pic/dog1.jpeg')
```

```python
features = {}
```

```python
features[2] = decode_img('pic/dog2.jpg')
features[3] = decode_img('pic/dog3.jpg')
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
f_v.reshape(1, -1).shape
```

```python
print(cosine_similarity(f_v.reshape(1, -1), features[2].reshape(1,-1)))
print(cosine_similarity(f_v.reshape(1, -1), features[3].reshape(1,-1)))
```
